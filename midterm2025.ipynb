ข้อมูลที่ได้จากบทสัมภาษณ์จากพยานทั้ง 2 คน
- คนร้ายเป็น ผู้ชาย

- เขาเป็นสมาชิกระดับ Gold ของยิม "Get Fit Now Gym"

- หมายเลขสมาชิกของเขาขึ้นต้นด้วย "48Z"

- เขาถูกพบเห็นที่ยิมในวันที่ 9 มกราคม

- เขาวิ่งหนีไปในรถที่มีป้ายทะเบียนซึ่งมีตัวอักษร "H42W"



```python
# Filter Get Fit Now members for Gold status and membership ID starting with '48Z'
gold_members_48Z = get_fit_now_member[(get_fit_now_member['membership_status'] == 'gold') & (get_fit_now_member['id'].str.startswith('48Z'))]

# Filter Get Fit Now check-ins for the date 20180109
checkins_jan_9 = get_fit_now_check_in[get_fit_now_check_in['check_in_date'] == 20180109]

# Merge the filtered membership and check-in dataframes
potential_killers_gym = pd.merge(gold_members_48Z, checkins_jan_9, left_on='id', right_on='membership_id')

# Get the person information for these potential killers
potential_killers_person = pd.merge(potential_killers_gym, person, left_on='person_id', right_on='id')

# Filter driver's licenses for plates containing 'H42W'
plates_H42W = drivers_license[drivers_license['plate_number'].str.contains('H42W')]

# Merge the potential killers with the driver's license information
final_suspects = pd.merge(potential_killers_person, plates_H42W, left_on='license_id', right_on='id')
display(final_suspects)
```

# **จากการให้การของพยายนชี้ว่า Jeremy Bowers เป็นคนร้ายในคดี**
และจากคำให้การของ Jeremy Bowers

"ฉันถูกผู้หญิงที่มีเงินเยอะจ้างมา ฉันไม่รู้ชื่อเธอ แต่รู้ว่าเธอสูงประมาณ 5'5" (65") or 5'7" (67"). เธอผมแดงและขับรถ Tesla Model S ฉันรู้ว่าเธอเคยไปดูคอนเสิร์ต SQL Symphony ถึง 3 ครั้งในเดือนธันวาคม 2017"

**ข้อมูลผู้บงการ/จ้างวาน : ผู้หญิง, ผมแดง, สูงประมาณ 165-170 ซม. (5'5" - 5'7"), ขับรถ Tesla Model S, เคยชมคอนเสิร์ต SQL Symphony 3 ครั้งในเดือน ธ.ค. 2017**

```python
# กรองจาก facebook_event_checkin คนที่ เคยชมคอนเสิร์ต SQL Symphony 3 ครั้งในเดือน ธ.ค. 2017
# Filter for check-ins at 'SQL Symphony Concert' in December 2017
sql_symphony_checkins_dec_2017 = facebook_event_checkin[(facebook_event_checkin['event_name'] == 'SQL Symphony Concert') & (facebook_event_checkin['date'] >= 20171201) & (facebook_event_checkin['date'] <= 20171231)]

# Count the number of SQL Symphony check-ins for each person
sql_symphony_counts = sql_symphony_checkins_dec_2017.groupby('person_id').size().reset_index(name='checkin_count')

# Filter for people who attended the concert exactly 3 times
sql_symphony_3_times = sql_symphony_counts[sql_symphony_counts['checkin_count'] == 3]

# Display the result
display(sql_symphony_3_times)
```

```python
# Filter for females with red hair and height between 165 and 170 cm
potential_masterminds_person = person.merge(drivers_license, left_on='license_id', right_on='id').query("gender == 'female' and hair_color == 'red' and height >= 65 and height <= 67")

# Filter for people who drive a Tesla Model S
tesla_drivers = drivers_license[(drivers_license['car_make'] == 'Tesla') & (drivers_license['car_model'] == 'Model S')]

# Merge potential masterminds with Tesla drivers
potential_masterminds_tesla = pd.merge(potential_masterminds_person, tesla_drivers, left_on='license_id', right_on='id')
potential_masterminds_tesla
```

```python
# Filter for females with red hair and height between 165 and 170 cm
potential_masterminds_person = person.merge(drivers_license, left_on='license_id', right_on='id').query("gender == 'female' and hair_color == 'red' and height >= 65 and height <= 67")

# Filter for people who drive a Tesla Model S
tesla_drivers = drivers_license[(drivers_license['car_make'] == 'Tesla') & (drivers_license['car_model'] == 'Model S')]

# Merge potential masterminds with Tesla drivers
potential_masterminds_tesla = pd.merge(potential_masterminds_person, tesla_drivers, left_on='license_id', right_on='id')

# Filter for check-ins at 'SQL Symphony Concert' in December 2017
sql_symphony_checkins_dec_2017 = facebook_event_checkin[(facebook_event_checkin['event_name'] == 'SQL Symphony Concert') & (facebook_event_checkin['date'] >= 20171201) & (facebook_event_checkin['date'] <= 20171231)]

# Count the number of SQL Symphony check-ins for each person
sql_symphony_counts = sql_symphony_checkins_dec_2017.groupby('person_id').size().reset_index(name='checkin_count')

# Filter for people who attended the concert exactly 3 times
sql_symphony_3_times = sql_symphony_counts[sql_symphony_counts['checkin_count'] == 3]

# Merge potential masterminds with those who attended SQL Symphony 3 times
final_masterminds = pd.merge(potential_masterminds_tesla, sql_symphony_3_times, left_on='id_x', right_on='person_id')

# Display the final mastermind
display(final_masterminds)
```
