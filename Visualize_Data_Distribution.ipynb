````
stopandsearch_df_dropna = stopandsearch_df.dropna(subset=['Age range','Gender'])
````

````
stopandsearch_df_dropna['Gender'] = stopandsearch_df_dropna['Gender'].replace({'Male': 35, 'Female': 1})
stopandsearch_df_dropna['Age range'] = stopandsearch_df_dropna['Age range'].replace({'under 10': 10, '10-17': 14, '18-24':21, '25-34':30, 'over 34':34})

display(stopandsearch_df_dropna.head())
````
