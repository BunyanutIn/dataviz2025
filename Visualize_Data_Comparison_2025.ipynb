```python
def swap_rows(tab, eth1='Asian', eth2='Other'):
    # Find the indices of 'Asian' and 'Other'
    asian_index = tab[tab['Officer-defined ethnicity'] == eth1].index[0]
    other_index = tab[tab['Officer-defined ethnicity'] == eth2].index[0]

    # Get the rows
    asian_row = tab.loc[asian_index].copy()
    other_row = tab.loc[other_index].copy()

    # Swap the rows
    tab.loc[asian_index] = other_row
    tab.loc[other_index] = asian_row

    return tab
```

```python
compared_police_eth1 = swap_rows(compared_police_eth1)
compared_police_eth2 = swap_rows(compared_police_eth2)
compared_police_eth3 = swap_rows(compared_police_eth3)
compared_police_eth4 = swap_rows(compared_police_eth4)
```

```python
import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย
```

```python
data = {'client_id' : ['c1' , 'c2' , 'c3' , 'c4'],
         'cheese' : [2,5,4,3],
         'beer' : [2,1,3,5],
         'chicken' : [5,3,3,4],
        'bread' : [4,3,2,1],
       'coffee' : [2,3,5,3]}
df = pd.DataFrame.from_dict(data)
df
```

```python
categories = list(df.columns[1:])
categories
```

```python
values = df.iloc[:,1:].mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

ax.plot(angles, values, linewidth=1, linestyle='solid' , color = '#ea3033') #เส้นกราฟ
ax.fill(angles, values, 'lightpink', alpha=0.5) #พื้นที่
```

```python
# 1. สร้าง Dictionary สำหรับการแปลงค่าเก็บไว้ในตัวแปร
age_replacement_map = {
    '10-17': 14,
    '18-24': 21,
    '25-34': 30,
    'over 34': 34,
    'under 10': 10
}


# 2. ใช้ for loop เพื่อสั่งให้ทำงานกับ DataFrame ทุกตัวที่ต้องการ
for df in [data1, data2, data3, data4]:
    df['Age range'] = df['Age range'].replace(age_replacement_map)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi


# ตัวอย่าง: หาค่าเฉลี่ยอายุของแต่ละเชื้อชาติ
age_by_eth = data1.groupby("Officer-defined ethnicity")["Age range"].mean()


categories = list(age_by_eth.index)
values = age_by_eth.values.flatten().tolist()
values += values[:1]  # วงกลมต้องปิด loop


# สร้างมุมรอบวงกลม
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]


# plot radar
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, color='grey', size=12)


ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, 'skyblue', alpha=0.4)


plt.title("Average Age (by Officer-defined ethnicity)", size=15, y=1.1)
plt.show()
```

```python
data1['Age range'] = pd.to_numeric(data1['Age range'], errors='coerce')
data2['Age range'] = pd.to_numeric(data2['Age range'], errors='coerce')


# ค่าเฉลี่ยอายุของแต่ละเชื้อชาติ
age_month1 = data1.groupby("Officer-defined ethnicity")["Age range"].mean()
age_month2 = data2.groupby("Officer-defined ethnicity")["Age range"].mean()


# รวมเป็น DataFrame เดียว
df_age = pd.DataFrame({
    "Month1": age_month1,
    "Month2": age_month2
})


# แกนรอบวงกลม
categories = list(df_age.index)
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]


# ค่าแต่ละเดือน
values1 = df_age["Month1"].tolist()
values1 += values1[:1]


values2 = df_age["Month2"].tolist()
values2 += values2[:1]


# สร้าง Radar Chart
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, size=12, color="grey")
ax.set_rlabel_position(30)
plt.yticks([10,15,20,25,30,35], ["10","15","20","25","30","35"], color="grey", size=10)
plt.ylim(10,35)


# plot Month1
ax.plot(angles, values1, linewidth=2, linestyle='solid', label="Month1")
ax.fill(angles, values1, 'skyblue', alpha=0.4)


# plot Month2
ax.plot(angles, values2, linewidth=2, linestyle='solid', label="Month2")
ax.fill(angles, values2, 'lightpink', alpha=0.4)


plt.title("Average Age Range by Ethnicity (Month1 vs Month2)", size=14, y=1.1)
plt.legend(loc="upper right", bbox_to_anchor=(0.1, 0.1))
plt.show()
```

```python
data_india = data_india.dropna(subset=['Rating','Duration','Genre'])
data_india = data_india[['Rating','Duration','Genre']]
data_india
```

```python
# prompt: แปลงค่าใน column duration เป็น int โดยลบ min ออกและ เปลี่ยนชนิดข้อมูลเป็น integer

# Remove " min" from the 'Duration' column and convert to integer
data_india['Duration'] = data_india['Duration'].str.replace(' min', '').astype(int)
```

```python
data_india.describe()
```

```python
# prompt: แยกตารางออกเป็นตารางย่อย 5 ตารางสำหรับ genre: drama, action, romance, horror, musical

# Create a dictionary to store the tables
genre_tables = {}

# Define the genres to extract
genres = ['Drama', 'Action', 'Romance', 'Horror', 'Musical']

# Iterate through the genres and create separate tables
for genre in genres:
  genre_tables[genre] = data_india[data_india['Genre'] == genre].copy()

# Access the tables using the genre as the key
# Example:
# drama_table = genre_tables['Drama']
```

```python
all_table = pd.concat(genre_tables.values())
all_table
```

```python
# prompt: แบ่ง duration ออกเป็น short, medium, long โดย short สั้นกว่า 60 medium 60-90 ที่เหลือ เป็น long

# Function to categorize duration
def categorize_duration(duration):
  if duration < 60:
    return 'short'
  elif 60 <= duration <= 90:
    return 'medium'
  else:
    return 'long'

# Apply the function to create a new column
all_table['Duration Category'] = all_table['Duration'].apply(categorize_duration)
all_table
```
